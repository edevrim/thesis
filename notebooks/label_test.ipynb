**Label graph**

```
# 0: subtour elimination
# 1: blossoms (basic combs)
# 2: advanced comb
# 3: clique tree
# 4: blossom + path 
# 41: pure path 
# 5: bipartition
# 6: envelope
# 7: crown 8 
# 8: crown multiple 
# 9: none working, branching needed 
def label_me(problem, cycle_cutoff): 
   
    subtour_flag = 0.0
    blossom_flag = 0.0 
    adv_comb_flag = 0.0 
    clique_tree_flag = 0.0 
    blossom_path_flag = 0.0 
    _path_flag = 0.0 
    bipartition_flag = 0.0 
    envelope_flag = 0.0
    crown8_flag = 0.0
    crownN_flag = 0.0
    
    subtour_delta = 0.0
    blossom_delta = 0.0 
    adv_comb_delta = 0.0 
    clique_tree_delta = 0.0 
    blossom_path_delta = 0.0 
    _path_delta = 0.0
    bipartition_delta = 0.0 
    envelope_delta = 0.0
    crown8_delta = 0.0
    crownN_delta = 0.0

    for action in range(9): 
        problem, change_flag, count_, delta_obj = act.solve_action_check(problem, action, cycle_cutoff)
        if change_flag == 1: 
          if action == 0:
            subtour_flag = count_ 
            subtour_delta = delta_obj
          if action == 1:
            blossom_flag = count_
            blossom_delta = delta_obj
          elif action == 2:
            adv_comb_flag = count_
            adv_comb_delta = delta_obj
          elif action == 3:
            clique_tree_flag = count_
            clique_tree_delta = delta_obj
          elif action == 4: 
            blossom_path_flag = count_
            blossom_path_delta = delta_obj
            if count_ > 1: 
              _path_flag = count_
              _path_delta = delta_obj
          elif action == 5: 
            bipartition_flag = count_
            bipartition_delta = delta_obj 
          elif action == 6: 
            envelope_flag = count_  
            envelope_delta = delta_obj 
          elif action == 7: 
            crown8_flag = count_      
            crown8_delta = delta_obj
          elif action == 8: 
            crownN_flag = count_      
            crownN_delta = delta_obj
    
    branch_flag = (0.0 if (subtour_flag + blossom_flag + adv_comb_flag + clique_tree_flag + blossom_path_flag + bipartition_flag + envelope_flag + crown8_flag + crownN_flag) > 0 else 1.0)
    
    label_ = [subtour_flag, blossom_flag, adv_comb_flag, clique_tree_flag, blossom_path_flag, bipartition_flag, envelope_flag, crown8_flag, crownN_flag, branch_flag, _path_flag]
    delta_ = [subtour_delta, blossom_delta, adv_comb_delta, clique_tree_delta, blossom_path_delta, bipartition_delta, envelope_delta, crown8_delta, crownN_delta, _path_delta]
    # return labels and delta
    return label_, delta_
```

**Collect graphs and labels**

```
def collect_graphs(RUN_NO, train_states, delta_states, tsp_id_list, global_count, len_train_states, final_objective_list):
 
  print("Saving now:", global_count, "Length of states:", len_train_states)
  #graphs = [item[0] for item in train_states]
  #labels = [item[0] for item in train_states]
  labels = train_states

  #save_graphs("train_graphs"+str(RUN_NO)+".bin", graphs)

  labels_pd = pd.DataFrame(labels)
  labels_pd['tsp_id'] = tsp_id_list
  labels_pd['objective'] = final_objective_list
  labels_pd.to_csv("TEST_LABELS"+str(RUN_NO)+".csv")

  delta_pd = pd.DataFrame(delta_states)
  delta_pd['tsp_id'] = tsp_id_list
  delta_pd.to_csv("TEST_DELTA"+str(RUN_NO)+".csv")

  files.download("TEST_LABELS"+str(RUN_NO)+".csv")
  files.download("TEST_DELTA"+str(RUN_NO)+".csv")
  #files.download("train_graphs"+str(RUN_NO)+".bin")

  return RUN_NO
```

```
train_states = []
delta_states = [] 
tsp_id_list = []
final_objective_list = []

RUN_NO = 7

# 1: only subtours 
# 2: subtours + blossom 
# 3: subtour + adv comb + blossom 
# 4: subtour + clique tree + blossom + adv comb
# 5: subtour + path 
# 6: bipartition + subtour + clique tree + blossom + adv comb 
# 7: crown8 + crownN + subtour 

len_train_states = 0
verbose1 = 0
global_count = 0 
cycle_cutoff = 0.99 

max_of_actions_per_tsp = 30 # max number of actions can be taken per tsp 
```

```
for k in range(list_len): 
    print("running iteration:", k)

    tsp_id = train_tsp_list[k][0]
    tsp_data = train_tsp_list[k][1]

    action_count = 0 
    print('TEST TSP_id:', tsp_id, 'Global_count:', global_count, '# of graphs gathered:', len_train_states)
    
    # init the problem
    problem1 = act.init_prob(tsp_id, tsp_data, verbose1, 'continuous')
    initial_objective = problem1.objective_val 
    
    # initial label 
    label, delta = label_me(problem1, cycle_cutoff)
    new_state = problem1.graph 
    new_state.add_edges_from(zip(new_state.nodes(), new_state.nodes()))
    new_state = dgl.DGLGraph(new_state)
             
    #train_states.append([new_state, label])
    train_states.append(label)
    delta_states.append(delta)
    tsp_id_list.append(tsp_id)
    final_objective_list.append(problem1.objective_val)
    old_state = new_state
    change_flag = 1 

    while change_flag == 1 and problem1.complete_flag == 0 and action_count <= max_of_actions_per_tsp:
        
        global_count = global_count + 1 
        action_count = action_count + 1 
        
        #save training graphs to keep saving
        len_train_states = len(train_states)
        if len_train_states > 0 and len_train_states%500 == 0:
          RUN_NO = collect_graphs(RUN_NO, train_states, delta_states, tsp_id_list, global_count, len_train_states, final_objective_list)  
        
        # update problem 
        #if label[0] > 0.0 and label[1] > 0.0:
        #  action = random.randint(0, 1)
        #if label[0] > 0.0: 
        #  action = 0
        if label[0] > 0.0: 
          action = 0 
        elif label[8] > 0.0: 
          action = 8
        #elif label[1] > 0.0: 
        #  action = 1      
        else: 
          action = 7
        #elif label[5] > 0.0:
        #    action = 5
        #elif label[2] > 0.0:
        #    action = 2 
        #elif label[3] > 0.0:
        #    action = 3
        #elif label[4] > 0.0:
        #    action = 4
        #else: 
        #    action = random.randint(9, 11)

        problem1, delta_obj, change_flag, new_state, count_const = act.solve_action_update(problem1, action, initial_objective, old_state, cycle_cutoff)       

        # add to training set if the problem has changed
        if change_flag == 1 and problem1.complete_flag == 0: 
            label, delta = label_me(problem1, cycle_cutoff)
            #train_states.append([new_state, label])
            train_states.append(label)
            delta_states.append(delta)
            tsp_id_list.append(tsp_id)
            final_objective_list.append(problem1.objective_val)
            old_state = new_state
```
